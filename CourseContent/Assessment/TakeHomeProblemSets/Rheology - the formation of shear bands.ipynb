## Non-linear feedbacks in the material strength


We have discussed what happens when the strength of a material under loading decreases as a function of the loading. Materials with this property often form highly localised regions of rapid deformation surrounded by stagnant regions where there is a very much lower rate of deformation.

In many cases, the 'failure' of one part of the system will unload / de-stress the remainder of the system and the  response to loading occurs as if the entire material had the lower strength of the failed regions. This is what leads to the flat parts of the stress v. strain or stress v. strain-rate curves from the lecture notes.

In this exercise you have been provided with two notebooks:

   - ShearBandsSimpleShear 
   - ShearBandsPureShear
   
The first of these is a wrapped-around-box which is sheared by top and bottom velocities. The box is full of a material with strength that is limited by friction:
\begin{equation}
    \eta =  
        \begin{cases}
              \eta_0 & |\tau| < \tau_\textrm{yield} \\
              {\tau_\textrm{yield}} / {2 \left| \, \dot\varepsilon \, \right|} & \textrm{otherwise}
        \end{cases}
\end{equation}
where the yield stress, $\tau_\textrm{yield}$ is given by

\begin{equation}
    \tau_\textrm{yield} = C(\varepsilon_p) + \mu p 
\end{equation}

$\left| \, \dot\varepsilon \, \right|$ is the second invariant of the strain rate tensor, $\varepsilon$ is its integral over time in regions where the yield stress is reached, $C$ is a cohesion, $\mu$ is a friction coefficient, and $p$ is the pressure.



The cohesion weakens (with accumulated plastic strain 
$\varepsilon_\textrm{p}$ ) 
as follows:

\begin{equation}
    C = c_0 + c_1 e^{ \left( -\varepsilon_\textrm{p} / \varepsilon_0 \right)}
\end{equation}


The second notebook is a _pure shear_ example in which the box is stretched at the sides and extends the sample. The material forms "shear bands" which are a little bit like normal faults.
